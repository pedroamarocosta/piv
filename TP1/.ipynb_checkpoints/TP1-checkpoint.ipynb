```python
def labeling(bwImage):
    if cv2.__version__.startswith('2.'):
        bw = bwImage.copy()
        (contourSeq, contourHierarchy) = cv2.findContours(bw, \
                    cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
    if cv2.__version__.startswith('3.'):
        (bw, contourSeq, contourHierarchy) = cv2.findContours(bwImage, \
                    cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
    if cv2.__version__.startswith('4.'):
        (contourSeq, contourHierarchy) = cv2.findContours(bwImage, \
                    cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
    grayLevel = 0
    outImage = np.zeros(bwImage.shape)
    idx = 0
    while idx >= 0:
        grayLevel = grayLevel+1
        color = (grayLevel,grayLevel,grayLevel)
        cv2.drawContours(outImage, contourSeq, idx, color, -1, 8 , contourHierarchy)
        idx = contourHierarchy[0][idx][0]

    return grayLevel+1, outImage
```

```python
#Verifica se determinado centro esta dentro de outro objecto
def isInsideAnother(allContours, thisCenter, thisRadius, idx, thisArea):
    #ve todos os objectos, e verifica o seu centro e raio para verificar se este centre esta dentro dele
    for i in range(len(contours)):
        #exclui o proprio
        if i == idx:
            continue

        contour = cntsSorted[i]
        arcL = cv2.arcLength(contour, True)
        radius = arcL / (2 * np.pi)
        area = cv2.contourArea(contour)

        #exclui os que nao tem tamanhos das moedas
        if (arcL < 320 or arcL > 600):
            continue

        #excluir areas menores que o que esta a verificar
        if area < thisArea:
            continue

        M = cv2.moments(contour)
        cX = int(M["m10"] / M["m00"])
        cY = int(M["m01"] / M["m00"])

        #Se o raio deste for maior que a distancia entre centros + o raio do 2o, entao o 2o esta dentro do 1o
        betweenCenters = math.sqrt(math.pow(thisCenter[0]-cX, 2) + math.pow(thisCenter[1]-cY,2))
        #-10 para correcao de imagens proximas que nao sao ruido
        if radius-10 > (betweenCenters - thisRadius):
            return True
    return False
```

### Métricas de cálculo da distância


#### Distância euclidiana

```python
def euclidean_distance_1(u, v):
    return np.sqrt(np.sum((u-v)**2))
def euclidean_distance_2(u, v):
    return np.linalg.norm(u-v)
def euclidean_distance_3(u, v):
    return scipy.spatial.distance.euclidean(u, v)

# Testes

print(euclidean_distance_1(np.array([1, 1]), np.array([1, 1])))
print(euclidean_distance_1(np.array([9, -9]), np.array([9, -1])))
print(euclidean_distance_1(np.array([0, -9]), np.array([1, 0])))
print(euclidean_distance_2(np.array([76, -11]), np.array([3, 16])))
print(euclidean_distance_2(np.array([8, 5]), np.array([2, 4])))
print()
print(euclidean_distance_2(np.array([1, 1]), np.array([1, 1])))
print(euclidean_distance_2(np.array([9, -9]), np.array([9, -1])))
print(euclidean_distance_2(np.array([0, -9]), np.array([1, 0])))
print(euclidean_distance_2(np.array([76, -11]), np.array([3, 16])))
print(euclidean_distance_2(np.array([8, 5]), np.array([2, 4])))
print()

# Referência

from scipy.spatial.distance import euclidean
def euclidean_distance_3(u, v):
    return euclidean(u, v)

print(euclidean_distance_3(np.array([1, 1]), np.array([1, 1])))
print(euclidean_distance_3(np.array([9, -9]), np.array([9, -1])))
print(euclidean_distance_3(np.array([0, -9]), np.array([1, 0])))
print(euclidean_distance_3(np.array([76, -11]), np.array([3, 16])))
print(euclidean_distance_3(np.array([8, 5]), np.array([2, 4])))
```

#### Distância de Manhattan

```python
def manhattan_distance_1(u, v):
    return np.abs(u - v).sum(-1)

def manhattan_distance_2(u, v):
    return np.sum(np.abs(v-u), axis=0)

# Testes

print(manhattan_distance_1(np.array([1, 1]), np.array([1, 1])))
print(manhattan_distance_1(np.array([9, -9]), np.array([9, -1])))
print(manhattan_distance_1(np.array([0, -9]), np.array([1, 0])))
print(manhattan_distance_1(np.array([76, -11]), np.array([3, 16])))
print(manhattan_distance_1(np.array([8, 5]), np.array([2, 4])))
print()
print(manhattan_distance_2(np.array([1, 1]), np.array([1, 1])))
print(manhattan_distance_2(np.array([9, -9]), np.array([9, -1])))
print(manhattan_distance_2(np.array([0, -9]), np.array([1, 0])))
print(manhattan_distance_2(np.array([76, -11]), np.array([3, 16])))
print(manhattan_distance_2(np.array([8, 5]), np.array([2, 4])))
print()

# Referência

from scipy.spatial.distance import cityblock
def manhattan_distance_3(u, v):
    return cityblock(u, v)

print(manhattan_distance_3(np.array([1, 1]), np.array([1, 1])))
print(manhattan_distance_3(np.array([9, -9]), np.array([9, -1])))
print(manhattan_distance_3(np.array([0, -9]), np.array([1, 0])))
print(manhattan_distance_3(np.array([76, -11]), np.array([3, 16])))
print(manhattan_distance_3(np.array([8, 5]), np.array([2, 4])))
```

#### Distância de cosseno

```python
def cosine_distance_1(u, v):
    return np.dot(u, v)/(np.linalg.norm(u)*np.linalg.norm(v))

# Testes

print(cosine_distance_1(np.array([1, 1]), np.array([1, 1])))
print(cosine_distance_1(np.array([9, -9]), np.array([9, -1])))
print(cosine_distance_1(np.array([0, -9]), np.array([1, 0])))
print(cosine_distance_1(np.array([76, -11]), np.array([3, 16])))
print(cosine_distance_1(np.array([8, 5]), np.array([2, 4])))
print()

# Referência

from scipy.spatial.distance import cosine
def cosine_distance_2(u, v):
    return 1-cosine(u[:, np.newaxis], v[:, np.newaxis])

print(cosine_distance_2(np.array([1, 1]), np.array([1, 1])))
print(cosine_distance_2(np.array([9, -9]), np.array([9, -1])))
print(cosine_distance_2(np.array([0, -9]), np.array([1, 0])))
print(cosine_distance_2(np.array([76, -11]), np.array([3, 16])))
print(cosine_distance_2(np.array([8, 5]), np.array([2, 4])))
print()
```
